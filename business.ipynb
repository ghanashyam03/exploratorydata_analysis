```python
cat_cols=data.select_dtypes(include=['object']).columns
num_cols = data.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(data[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[col])
    plt.show()
```

```python

# Create a 3x2 grid of subplots
fig, axes = plt.subplots(4, 2, figsize=(18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')

# Create countplots for each categorical variable
for i, var in enumerate(cat_cols):
    
    row = i // 2  # Calculate the row index
    col = i % 2   # Calculate the column index
    sns.countplot(ax=axes[row, col], x=var, data=data, color='blue', order=data[var].value_counts().index)
    axes[row, col].tick_params(labelrotation=45) 
    if var in ['State']:
        axes[row, col].tick_params(labelrotation=90)  # Rotate x-axis labels by 90 degrees for better readability
        axes[row, col].tick_params(axis='x', labelsize=8) # Rotate x-axis labels for better readability

plt.tight_layout()  # Adjust subplot spacing
plt.show()
```

```python
top_cities = data['City'].value_counts().head(20)

# Get the worst 20 cities by frequency
worst_cities = data['City'].value_counts().tail(20)

# Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot the top 20 cities
sns.barplot(ax=axes[0], x=top_cities.values, y=top_cities.index)
axes[0].set_title('Top 20 Cities by Frequency')
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('City')
axes[0].tick_params(axis='y', labelrotation=0)  # Keep y-axis labels horizontal

# Plot the worst 20 cities
sns.barplot(ax=axes[1], x=worst_cities.values, y=worst_cities.index)
axes[1].set_title('Worst 20 Cities by Frequency')
axes[1].set_xlabel('Frequency')
axes[1].set_ylabel('City')
axes[1].tick_params(axis='y', labelrotation=0)  # Keep y-axis labels horizontal

plt.tight_layout()
plt.show()
```

```python

def log_transform(data, col):
    for colname in col:
        valid_values = data[data[colname] > 0][colname]  
        data[colname + '_log'] = np.log1p(valid_values) 
```

```python
data_filtered = data.dropna(subset=['Profit'])
log_transform(data_filtered, ['Sales', 'Profit'])
sns.displot(data=data_filtered, x="Sales_log", kde=True, bins=20)
plt.xlabel("Sales_log")
plt.ylabel("Density")
plt.title("Distribution of Sales_log")
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=data.drop(['Sales','Profit'],axis=1))
plt.show()
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

# Features to plot against 'Profit'
features_to_plot = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']

# Loop through each feature and create a bar plot
for i, feature in enumerate(features_to_plot):
    # Group by the feature and calculate the mean of 'Profit', sorted in descending order
    data.groupby(feature)['Profit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[i // 2][i % 2], fontsize=12)
    axarr[i // 2][i % 2].set_title(f"{feature} Vs Profit", fontsize=18)

# Adjust subplot spacing
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=0.5)

# Remove spines from plots
sns.despine()

# Show the plots
plt.show()
```

```python
state_profit_means = data.groupby('State')['Profit'].mean().sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x=state_profit_means.index, y=state_profit_means.values)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('State', fontsize=14)
plt.ylabel('Mean Profit', fontsize=14)
plt.title('Mean Profit by State', fontsize=18)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
top_cities = data.groupby('City')['Profit'].sum().sort_values(ascending=False).head(20)

# Create a bar plot for the top N cities
plt.figure(figsize=(12, 8))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Profit', fontsize=14)
plt.title('Top 20 Cities by Total Profit', fontsize=18)
plt.tight_layout()

# Show the plot
plt.show()
```
